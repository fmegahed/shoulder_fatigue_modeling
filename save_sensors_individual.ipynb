# Generate the Individual Models for Sensors

In the research data, there are three different sensors used to gather data - wrist, trunk and upperarm. Each of these sensors provides multiple different features that are used in generating the models. The purpose of generating models using only features form specific sensors is to see if some sensors are better for predicting fatigue than others.

Here, individual models are generated for each of the sensors. These models follow the same process as the individual models for all sensors, except the models are generated 3 times - once for each sensor.

The total number of models generated per sensor is 252 (as discussed in the `save_models_individual.ipynb` file). 252 × 3 = 756 total models generated.

The models are generated and then saved in a CSV file called `sensor_train_individual.csv` for the training models and `sensor_test_individual.csv` for the testing models. The feature importance data is also saved for each model. This data is saved in a CSV file called `sensor_feature_importance_general.csv`. These files are in the directory `outputs/individual/`.

```python
from pycaret.datasets import get_data
from pycaret.regression import *
import csv
import pandas as pd
```

```python
# SAVE THE NEW DFs AS CSVs
df_original = get_data('data/initial_features_mod')
df_wrist = df_original.filter(regex='wrist')
df_trunk = df_original.filter(regex='trunk')
df_upperarm = df_original.filter(regex='upperarm')
df_wrist[['subject', 'rpe']] = df_trunk[['subject', 'rpe']] = df_upperarm[['subject', 'rpe']] = df_original[['subject', 'rpe']].values
sensor_df_list = [df_wrist, df_trunk, df_upperarm]
sensor_ref = {
    0 : "wrist",
    1 : "trunk",
    2 : "upperarm"
}
```

```python
with open('outputs/individual/sensor_train_individual.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Model", "MAE", "MSE", "RMSE", "R2", "RMSLE", "MAPE", "TT (Sec)", "Subject", "Sensor"])

with open('outputs/individual/sensor_test_individual.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Model", "MAE", "MSE", "RMSE", "R2", "RMSLE", "MAPE", "Subject", "Sensor"])

with open('outputs/individual/sensor_feature_importance_individual.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Feature", "Value", "Model", "Subject", "Sensor"])

for index, df in enumerate(sensor_df_list):
    for subject in df['subject'].unique():
        # Splits the data 80 - 20
        train = df[df['subject'] == subject].sample(frac=0.8, random_state=42)
        test = df.drop(train.index)

        reg = setup(data=train, target='rpe', ignore_features=['experimental_condition', 'subject'])
        best = compare_models(sort='MAE', n_select = 18)
        all = pull()
        all['Subject'] = str(subject)
        all['Sensor'] = sensor_ref[index]

        all.to_csv('outputs/individual/sensor_train_individual.csv', mode='a', header=False, index=False)

        test_results = pd.DataFrame()
        for model in best:
            test_result = predict_model(model, verbose = False)
            test_result = pull()
            test_results = test_results.append(test_result)

            # NEW: MAKE SURE IT WORKS
            try:
                importance = pd.DataFrame({'Feature': get_config('X_train').columns, 'Value' : abs(model.feature_importances_)}).sort_values(by='Value', ascending=False).reset_index().drop('index', axis=1)
            except:
                try:
                    importance = pd.DataFrame({'Feature': get_config('X_train').columns, 'Value' : abs(model.coef_)}).sort_values(by='Value', ascending=False).reset_index().drop('index', axis=1)
                except:
                    importance = pd.DataFrame({'Feature': ['error'], 'Value': [0]})
            importance['Model'] = str(model)
            importance['Subject'] = str(subject)
            importance['Sensor'] = sensor_ref[index]
            importance.to_csv('outputs/individual/sensor_feature_importance_individual.csv', mode='a', header=False, index=False)


        test_results['subject'] = str(subject)
        test_results['Sensor'] = sensor_ref[index]
        test_results.to_csv('outputs/individual/sensor_test_individual.csv', mode='a', header=False, index=False)
```
