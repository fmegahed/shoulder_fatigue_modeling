```python
df_select = df.loc[
    (df['ConvertedCompYearly'] > 150000) & df['Country'].isin(['United States of America', 'United Kingdom of Great Britain and Northern Ireland']), 
       # Select rows
       ['ConvertedCompYearly', 'Age', 'Country', 'LanguageHaveWorkedWith', 'LearnCode']]
```

```python
df_select.loc[
    (df['LanguageHaveWorkedWith'].str.contains('Python', na=False)), 
       # Select rows
       ['ConvertedCompYearly', 'LanguageHaveWorkedWith']]
```

```python
df_select.sort_values(by=['Country', 'ConvertedCompYearly', 'Age'], ascending=[False, True, True])[
    # Reordering columns
    ['Country', 'ConvertedCompYearly', 'Age', 'LanguageHaveWorkedWith', 'LearnCode']
]

# Use inplace=True to set default order to sorted
# Then use df_select.sort_index() to sort the indexes to their proper order
```

```python
df_select_age_salary = df.loc[(df['Age'].notna()) & (df['ConvertedCompYearly'].notna()), ['Age', 'ConvertedCompYearly']]

grouped_df = df_select_age_salary.groupby('Age')['ConvertedCompYearly'].median().reset_index()

# create a bar plot from the DataFrame
grouped_df.plot(kind='bar', x='Age', y='ConvertedCompYearly', color=['aquamarine', 'lightsteelblue', 'cornflowerblue', 'slategrey'])

# set the axis labels and title
plt.xlabel('Age', fontsize=14)
plt.ylabel('Salary (USD)', fontsize=14)
plt.title('Age vs. Salary', fontsize=16)

# show the plot
plt.show()
```
