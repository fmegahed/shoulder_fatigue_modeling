# Generate the General Models for Sensors

In the research data, there are three different sensors used to gather data - wrist, trunk and upperarm. Each of these sensors provides multiple different features that are used in generating the models. The purpose of generating models using only features form specific sensors is to see if some sensors are better for predicting fatigue than others.

Here, general models are generated for each of the sensors. These models follow the same process as the general models for all sensors, except the models are generated 3 times - once for each sensor.

The total number of models generated per sensor is 6552 (as discussed in the `save_models_general.ipynb` file). 6552 × 3 = 19656 total models generated.

The models are generated and then saved in a CSV file called `sensor_train_general.csv` for the training models and `sensor_test_general.csv` for the testing models. The feature importance data is also saved for each model. This data is saved in a CSV file called `sensor_feature_importance_general.csv`. These files are in the directory `outputs/general/`.

```python
from pycaret.datasets import get_data
from pycaret.regression import *
from itertools import combinations
import csv
import pandas as pd
```

```python
df_original = get_data('data/initial_features_mod')
df_wrist = df_original.filter(regex='wrist')
df_trunk = df_original.filter(regex='trunk')
df_upperarm = df_original.filter(regex='upperarm')
df_wrist[['subject', 'rpe']] = df_trunk[['subject', 'rpe']] = df_upperarm[['subject', 'rpe']] = df_original[['subject', 'rpe']].values
sensor_df_list = [df_wrist, df_trunk, df_upperarm]
sensor_ref = {
    0 : "wrist",
    1 : "trunk",
    2 : "upperarm"
}


# The number of subjects used in testing data. Should be set to 3 to match the data used in results.
COMBINATIONS = 3


combinations_list = list(combinations(df_original['subject'].unique(), COMBINATIONS))
```

```python
with open('outputs/general/sensor_train_general.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Model", "MAE", "MSE", "RMSE", "R2", "RMSLE", "MAPE", "TT (Sec)", "Training Subjects", "Sensor", "Combinations"])

with open('outputs/general/sensor_test_general.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Model", "MAE", "MSE", "RMSE", "R2", "RMSLE", "MAPE", "Training Subjects", "Sensor", "Combinations"])

with open('outputs/general/sensor_feature_importance_general.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Feature", "Value", "Model", "Training Subjects", "Sensor", "Combinations"])


# Predictors to remove: subject
for index, df in enumerate(sensor_df_list):
    for sub in combinations_list:
        train = df[~df['subject'].isin(sub)]
        test = df[df['subject'].isin(sub)]

        reg = setup(data=train, target='rpe', ignore_features=['experimental_condition', 'subject'])
        best = compare_models(sort='MAE', n_select = 18)
        all = pull()
        all['test_set'] = str(sub)
        all['sensor'] = sensor_ref[index]
        all['combinations'] = COMBINATIONS

        all.to_csv('outputs/general/sensor_train_general.csv', mode='a', header=False, index=False)

        test_results = pd.DataFrame()
        for model in best:
            test_result = predict_model(model, verbose = False)
            test_result = pull()
            test_results = test_results.append(test_result)

            try:
                importance = pd.DataFrame({'Feature': get_config('X_train').columns, 'Value' : abs(model.feature_importances_)}).sort_values(by='Value', ascending=False).reset_index().drop('index', axis=1)
            except:
                try:
                    importance = pd.DataFrame({'Feature': get_config('X_train').columns, 'Value' : abs(model.coef_)}).sort_values(by='Value', ascending=False).reset_index().drop('index', axis=1)
                except:
                    importance = pd.DataFrame({'Feature': ['error'], 'Value': [0]})
            importance['Model'] = str(model)
            importance['test_set'] = str(sub)
            importance['sensor'] = sensor_ref[index]
            importance['combinations'] = COMBINATIONS
            importance.to_csv('outputs/general/sensor_feature_importance_general.csv', mode='a', header=False, index=False)

        test_results['test_set'] = str(sub)
        test_results['sensor'] = sensor_ref[index]
        test_results['combinations'] = COMBINATIONS
        test_results.to_csv('outputs/general/sensor_test_general.csv', mode='a', header=False, index=False)
    
```
